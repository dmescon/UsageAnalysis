```python
#define billing date issues
def billing_date_issues(df2):
    i = 0
    prevalues = df2.values[i]
    while i < df2.shape[0]:
        curvalues = df2.values[i]
        if curvalues[0] == prevalues[0] and i != 0: #same id
            if curvalues[3] < prevalues[4]: #StartDate before previous EndDate
                print(i,' ERROR: Prior EDate AFTER Current BDate:\nPre: ', prevalues, '\nCur: ', curvalues, '\n')
                prevalues = curvalues
                i += 1
            elif (curvalues[3] - prevalues[4]).days > 1:
                unbilleddays = (curvalues[3] - prevalues[4]).days
                print(i,' ERROR: Unbilled Days : ', unbilleddays, '\nPre: ', prevalues, '\nCur: ', curvalues, '\n')
                prevalues = curvalues
                i += 1
            elif curvalues[3] == prevalues[4]: 
                print(i,' ERROR: Prior EDate EQUALS Current BDate:\nPre: ', prevalues, '\nCur: ', curvalues, '\n')
                prevalues = curvalues
                i += 1        
            else:
                prevalues = curvalues
                i += 1
                continue
        else:
            prevalues = curvalues
            i += 1        
    return
```
